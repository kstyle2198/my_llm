```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the pre-trained model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# List of words to calculate similarity
words = ["photon", "physics", "peter rabbit", "bunny", "grass", "chocolate", "cookies"]
```

```python
# Encode words into embeddings
embeddings = model.encode(words)
embeddings.shape
```

```python
# Calculate cosine similarity matrix
similarity_matrix = cosine_similarity(embeddings)
```

```python
# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(similarity_matrix, annot=True, xticklabels=words, yticklabels=words, cmap="YlGnBu", cbar_kws={'label': 'Cosine Similarity'})
plt.title("Semantic Similarity Heatmap")
plt.show()
```
